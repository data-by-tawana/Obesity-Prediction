#**Data Preprocessing**

```python
# Check for missing values
print(df.isnull().sum())
```

```python
import numpy as np
#Separate features and the target
X = df.iloc[:,:-1].values  # Select all columns except the last one
Y = df.iloc[:, -1].values   # Select the last column
```

```python
print('Features :', X)
```

```python
print('Target :', Y)
```

```python
import sklearn
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[8,14,15])],remainder= 'passthrough')#to ensure other columns are kept
X = np.array (ct.fit_transform(X))
```

```python
print (X)
```

Splitting Features and Target

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size= 0.2,random_state= 1)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
print(X.shape)
```

Feature Scaling



```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,[1,2,3,6,7,8,10,12,14,15]] = sc.fit_transform(X_train[:,[1,2,3,6,7,8,10,12,14,15]])
X_test[:,[1,2,3,6,7,8,10,12,14,15]] = sc.fit_transform(X_test[:,[1,2,3,6,7,8,10,12,14,15]])
```

```python
print('Xtrain :', X_train)
```

```python
print('Xtest :' , X_test)
```

Feature Selection

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import f1_score
#Initialization
gbc = GradientBoostingClassifier(max_depth=5, random_state=1)
#Classifier using all features
gbc.fit(X_train, Y_train)
#Prediction creation
pred = gbc.predict(X_test)
#Model Evaluation
f1_score_all = round(f1_score(Y_test,pred, average = 'weighted'),3)

print(f1_score_all)
```

Variance Threshold

```python
#Variance of each feature
X_train.var(axis=0)
```

```python
#Making copies to avoid affecting original data
X_train_v1, X_test_v1, Y_train_v1, Y_test_v1 = X_train.copy(), X_test.copy(),Y_train.copy(), Y_test.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_X_train = scaler.fit_transform(X_train_v1)
print(scaled_X_train)
```

```python
column_names = ct.get_feature_names_out()
df_trans = pd.DataFrame(X, columns=column_names)

fig, ax = plt.subplots(figsize=(10,8))

x = df_trans.columns
y = scaled_X_train.var(axis=0)

plt.xticks(rotation=180, ha="right", fontsize=10)

ax.bar(x, y, width=0.2)
ax.set_xlabel('Features')
ax.set_ylabel('Variance')
ax.set_title('scaled features variance')
ax.set_ylim(0, 0.3)

for index, value in enumerate(y):
    plt.text(x=index, y=value+0.001, s=str(round(value, 3)), ha='center')

fig.autofmt_xdate()
plt.tight_layout()
```

```python
#my personal minimum threshold is 0.02
columns_kept = [col for col in column_names if col not in ['encoder__x14_Always', 'encoder__x15_Bike', 'encoder__x15_Motorbike']]

# Use boolean indexing on the columns
mask = np.isin(column_names, columns_kept)

# Apply the mask to select  columns
mask_X_train = X_train_v1[:, mask]#the scaled x train gives a variance of 0.293
mask_X_test = X_test_v1[:, mask]

gbc.fit(mask_X_train,Y_train_v1)
var_pred = gbc.predict(mask_X_test)
f1_score_var = round(f1_score(Y_test_v1,var_pred, average='weighted'),3)

print(f1_score_var)
```

```python
#comparison plot
fig, ax = plt.subplots()

x = ['features', 'Variance threshold']
y = [f1_score_all, f1_score_var]

ax.bar(x, y, width=0.2)
ax.set_xlabel('feature selection method')
ax.set_ylabel('weighted F1 Score')
ax.set_ylim(0, 1.2)

for index, value in enumerate(y):
    plt.text(x=index, y=value+0.01, s=str(round(value, 3)), ha='center')

plt.tight_layout()
```
